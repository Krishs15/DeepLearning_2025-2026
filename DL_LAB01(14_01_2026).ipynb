##Upload in colab folder

```python
import numpy as np
import os
import struct

# Define the helper function to load IDX files
def load_idx_file(filepath):
    """Loads IDX file (e.g., MNIST images or labels) and returns as a NumPy array."""
    with open(filepath, 'rb') as f:
        magic = struct.unpack('>I', f.read(4))[0]
        if magic == 2051: # Images
            num_items = struct.unpack('>I', f.read(4))[0]
            num_rows = struct.unpack('>I', f.read(4))[0]
            num_cols = struct.unpack('>I', f.read(4))[0]
            data = np.frombuffer(f.read(), dtype=np.uint8)
            return data.reshape(num_items, num_rows, num_cols)
        elif magic == 2049: # Labels
            num_items = struct.unpack('>I', f.read(4))[0]
            data = np.frombuffer(f.read(), dtype=np.uint8)
            return data.reshape(num_items)
        else:
            raise ValueError(f"Invalid magic number {magic} in {filepath}")

# Define file paths for all MNIST dataset components
train_images_path = '/content/train-images.idx3-ubyte'
train_labels_path = '/content/train-labels.idx1-ubyte'
test_images_path = '/content/t10k-images.idx3-ubyte'
test_labels_path = '/content/t10k-labels.idx1-ubyte'

# --- Load and display metadata for Training Images ---
print("--- Metadata for Training Images ---")
if os.path.exists(train_images_path):
    train_images = load_idx_file(train_images_path)
    print(f"File: {os.path.basename(train_images_path)}")
    print(f"  Shape: {train_images.shape}")
    print(f"  Data type: {train_images.dtype}")
    print(f"  Number of images: {train_images.shape[0]}")
    print(f"  Image dimensions: {train_images.shape[1]}x{train_images.shape[2]}")
    print(f"  Min pixel value: {np.min(train_images)}")
    print(f"  Max pixel value: {np.max(train_images)}")
else:
    print(f"Error: Training images file not found at {train_images_path}.")

print("\n" + "-"*40 + "\n")

# --- Load and display metadata for Training Labels ---
print("--- Metadata for Training Labels ---")
if os.path.exists(train_labels_path):
    train_labels = load_idx_file(train_labels_path)
    print(f"File: {os.path.basename(train_labels_path)}")
    print(f"  Shape: {train_labels.shape}")
    print(f"  Data type: {train_labels.dtype}")
    print(f"  Number of labels: {train_labels.shape[0]}")
    print(f"  Unique labels: {np.unique(train_labels)}")
    print(f"  Label counts: {np.bincount(train_labels)}")
else:
    print(f"Error: Training labels file not found at {train_labels_path}.")

print("\n" + "-"*40 + "\n")

# --- Load and display metadata for Test Images ---
print("--- Metadata for Test Images ---")
if os.path.exists(test_images_path):
    test_images = load_idx_file(test_images_path)
    print(f"File: {os.path.basename(test_images_path)}")
    print(f"  Shape: {test_images.shape}")
    print(f"  Data type: {test_images.dtype}")
    print(f"  Number of images: {test_images.shape[0]}")
    print(f"  Image dimensions: {test_images.shape[1]}x{test_images.shape[2]}")
    print(f"  Min pixel value: {np.min(test_images)}")
    print(f"  Max pixel value: {np.max(test_images)}")
else:
    print(f"Error: Test images file not found at {test_images_path}.")

print("\n" + "-"*40 + "\n")

# --- Load and display metadata for Test Labels ---
print("--- Metadata for Test Labels ---")
if os.path.exists(test_labels_path):
    test_labels = load_idx_file(test_labels_path)
    print(f"File: {os.path.basename(test_labels_path)}")
    print(f"  Shape: {test_labels.shape}")
    print(f"  Data type: {test_labels.dtype}")
    print(f"  Number of labels: {test_labels.shape[0]}")
    print(f"  Unique labels: {np.unique(test_labels)}")
    print(f"  Label counts: {np.bincount(test_labels)}")
else:
    print(f"Error: Test labels file not found at {test_labels_path}.")
```
