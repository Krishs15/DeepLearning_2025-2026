```python
#Importing Libraries
import numpy as np
import matplotlib.pyplot as plt

# 1. Prepare the Data
# X (feature)
X = np.array([1, 2, 3, 4, 5, 6, 7, 8])
# y (target)
y = np.array([2, 3, 5, 7, 8, 10, 12, 13])

# 2. Calculate "Best Fit" using OLS Formulas
# Step 2a: Calculate the means (averages) of X and y
mean_x = np.mean(X)
mean_y = np.mean(y)

# Step 2b: Calculate the two parts of the slope formula (numerator and denominator)
numerator = np.sum((X - mean_x) * (y - mean_y))
denominator = np.sum((X - mean_x)**2)

# Step 2c: Calculate the slope (m)
m = numerator / denominator

# Step 2d: Calculate the intercept (b)
b = mean_y - (m * mean_x)

#3. Show the Results
print("'Best Fit' (OLS) Results")
print(f"Calculated Slope (m): {m:.2f}")
print(f"Calculated Intercept (b): {b:.2f}")
print(f"The equation of the line: y = {m:.2f}x + {b:.2f}")

# 4. Visualize the Results
# Calculate the predicted 'y' values using our new m and b
y_predicted = (m * X) + b

# Plot the original data points
plt.scatter(X, y, color='blue', label='Actual Data')

# Plot our new "best fit" line
plt.plot(X, y_predicted, color='red', linewidth=2, label='Best Fit Line')

plt.title("Linear Regression(OLS)")
plt.xlabel('X (Feature)')
plt.ylabel('y (Target)')
plt.legend()
plt.grid(True)
plt.show()
```
